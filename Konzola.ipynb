```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def monte_carlo(n):
    L = np.random.normal(2.0, 0.02, n)
    P = np.maximum(np.random.normal(2000, 400, n), 1)
    E = np.random.normal(210e9, 5e9, n)
    b = np.random.uniform(0.05, 0.12, n)
    h = np.random.uniform(0.02, 0.05, n)
    I = b * h**3 / 12
    delta = P * L**3 / (3 * E * I)
    return pd.DataFrame({"L":L,"P":P,"E":E,"b":b,"h":h,"delta":delta})

for N in [100, 1000, 10000]:
    df = monte_carlo(N)
    print(f"\nN = {N}")
    print("min =", df["delta"].min())
    print("max =", df["delta"].max())
    print("mean =", df["delta"].mean())
    print("std =", df["delta"].std())
    plt.hist(df["delta"], bins=30)
    plt.title(f"Histogram  pre {N} realizci")
    plt.xlabel(" [m]")
    plt.ylabel("Poet vskytov")
    plt.show()
```
